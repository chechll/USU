## The `numpy` Package

We have already mentioned the `numpy` package in an earlier notebook. It is one of the most commonly used Python packages. It provides n-dimensional arrays and matrices and implements basic mathematical operations on them, along with many other functions.

### Creating a New `numpy` Array

When using `numpy`, the first step is, of course, to import it — typically using the alias `np`. After importing, you can create an array with the `array()` function, which accepts a Python list (for example) as an argument. For example:


````python
# @title
import numpy as np

x = np.array([1, 2.5, 3])

print(x)
````

Using a nested list, you can create a two dimensional array

````python
A = np.array([[1, 2.5, 3 ], [4,  5,  6, ], [7,  8,  9]])
print(f"A:\n{A}")
````

### Creating Special Arrays

There are several built-in function, which enable easier creation of commonly used special arrays. The shape of the array is passed in a list: `[2, 2]`, or as a tuple: `(2, 2)`:



````python
# Creating an all-zeros array.
A = np.zeros((3, 4))
print("Array of zeros \n:A = \n{}\n".format(A))

# Creating an all-ones array.
B = np.ones((4, 4))
print("Array of ones \n:B = \n{}\n".format(B))

# Creating an array filled with a constant.
C = np.full((3, 4), 7)
print("Array filled with a constant \nC = \n{}\n".format(C))

# Creating a square matrix with ones at the diagonal and zeros elsewhere.
D = np.eye(4)
print("Identity matrix\n D = \n{}\n".format(D))
````

Arrays can also have more dimensions, e.g., the following array has 3 dimensions:



````python
A = np.zeros([2, 2, 3])
print(A)
````

NumPy provides functions for creating random arrays. For example, to generate a `3 × 3` array with uniformly distributed random values from the interval $[0, 5)$, you can use the random number generator API. Note that the shape is passed as a tuple to the `size` (or `shape`) parameter.


````python
A = np.random.uniform(0, 5, (3, 3))
print(A)
````

### Indexing Elements of an Array

The individual elements of an array are indexed by listing their coordinates separated by commas in square brackets:



````python
A = np.array([[1, 2, 3], [4, 5, 6]])
print("A = \n{}\n".format(A))

a = A[0, 0]
print("A[0, 0] = {}".format(a))

b = A[0, 1]
print("A[0, 1] = {}".format(b))
````

A colon `:` can be used to select (slice) a segment of elements along a given axis (dimension):


````python
A = np.array([[1, 2, 3], [4, 5, 6]])
print("A = \n{}\n".format(A))

a = A[:2, 1]
print("A[:2, 1] = {}\n".format(a))

b = A[:, 0]
print("A[:, 0] = {}\n".format(b))

c = A[:, 1:3]
print("A[:, :2] = \n{}".format(c))
````

Similar to Python lists, NumPy arrays support indexing from the end using **negative indices** (e.g. `a[-1]` is the last element).

The colon `:` denotes the **slicing operator**, which behaves similarly to the `range()` function. The syntax is:

* ``a[start:stop]`` — elements from `start` up to **but not including** `stop`  
* ``a[start:]`` — elements from `start` through to the end of the array  
* ``a[:stop]`` — elements from the beginning up to **but not including** `stop`  
* ``a[:]`` — a slice that covers the entire array (in NumPy this usually returns a *view*, not a copy)

You can also specify a step value with any of the above forms:

* ``a[start:stop:step]`` — elements from `start` up to (but not including) `stop`, taken every `step` elements


### The Shape of an Array

Each numpy array has a shape which can be determined using attribute `.shape`:



````python
a = np.arange(1, 7) # array of integers from 1 to 6
print(f"Shape: {a.shape}")
````

Here, `a` is a **one-dimensional array**, so its `shape` is `(6,)` — a single-element tuple whose only entry is the length `6` (the trailing comma indicates a one-element tuple).

To create an explicit **column** or **row** vector (2-D arrays with shapes `(6, 1)` and `(1, 6)` respectively) you can use `reshape()`:


````python
a_column = a.reshape(2,3)
print(f"column vector:\n {a_column} \n with a shape: {a_column.shape}")
````

A more convenient way to convert a one-dimensional vector into a column or row vector ( $(n × 1)$ or $(1 \times n)$ matrix) is to pass `-1` as one of the arguments to the `reshape()` method. The `-1` tells NumPy to infer that dimension automatically while you explicitly set the other to `1`. So, to create a column vector, we can pass `-1` and `1` as arguments, so that the number of rows will be the size of the original vector.


````python

b = a.reshape(-1,1)
print(f"b:\n {b} \nshape: {b.shape}")

c = a.reshape(1,-1)
print(f"c:\n {c} \nshape: {c.shape}")
````

Also, we can use reshape to change the dimensions, e.g. to create a $2 \times 3$ matrix form the vector `a`. In the second example we create matrix with 3 rows, and the number of columns will be inferred.

You can also use `reshape()` to change an array's dimensions. For example, to convert a 1-D vector `a` of length 6 into a \(2\times 3\) matrix, call:


````python
D = a.reshape(2, 3)
print(f"D:\n {D}\n shape: {D.shape}")

# 3 x 2 matrix
E = a.reshape((3, -1))
print(f"E:\n {E}\n shape: {E.shape}")
````

More examples of applying reshape to $3$ dimensional arrays.

````python
A = np.zeros([2, 2, 3])
print(f"A.shape: {A.shape}")

B = np.reshape(A, [3, 2, 2])
print(f"B.shape: {B.shape}")

# The size in one dimension can be computed automatically,
# if we enter -1 instead of specifying the dimension.
C = np.reshape(A, [3, -1, 2])
print(f"C.shape: {C.shape}")
````

An array can be transposed using function `.transpose` or just by calling the attribute `T`:



````python
A = np.array([[1, 2, 3], [4, 5, 6]])
print("A = \n{}\n".format(A))

ATr = A.transpose()
print("ATr = \n{}\n".format(ATr))

AT = A.T
print("AT = \n{}\n".format(AT))
````

### Arithmetic Operations with Arrays

Arrays support the basic arithmetic operations, such as addition or multiplication which are applied element-wise:



````python
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[6, 5, 4], [3, 2, 1]])
print("A = \n{}\n".format(A))
print("B = \n{}\n".format(B))

# Element-wise addition:
C = A + B
print(f"A + B = \n{C}\n")

# Element-wise subtraction:
C = A - B
print(f"A - B = \n{C}\n")

# Element-wise multiplication:
C = A * B
print(f"A * B = \n{C}\n")

# Element-wise division:
C = A / B
print(f"A / B = \n{C}\n")
````

The `numpy` package also contains version of some standard mathematical function, such as `sin`, `cos` or `exp`, which can operate on arrays:



````python
A = np.array([[1, 2, 3], [4, 5, 6]])
print("A = \n{}\n".format(A))

sinA = np.sin(A)
print(f"sin(A) = \n{sinA}\n")

expA = np.exp(A)
print(f"exp(A) = \n{expA}\n")
````

### Algebraic Operations

The standard algebraic operations are also defined for `numpy` arrays, including elementwise arithmetic, dot products, and matrix multiplication.

````python
a = np.array([[1], [2], [3]])
print(f"a = {a}\n")
b = np.array([[4], [5], [6]])
print(f"b = {b}\n")

c = np.dot(a.T, b)
print(f"a^T.b =\n{c}")

# or alternatively using matrix multiplication operator @
c = a.T @ b
print(f"a^T.b =\n{c}")
````

````python
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[6, 5, 4], [3, 2, 1]])
print(f"A = \n{A}\n")
print(f"B = \n{B}\n")

C = np.matmul(A, B.transpose())
print(f"A x B = \n{C}")
````

Matrix multiplication actually also has a dedicated operator `@`, which can be used instead of `matmul` for better readability.



````python
C = A @ B.transpose()
print(f"A x B^T = \n{C}")
````

### Group Comparisons and Indexing

You can perform elementwise comparisons on a `numpy` array — for example, test which elements are greater than or equal to `5`. The result is a **boolean array** of the same shape, where each entry indicates whether the condition holds for the corresponding element. You can use that boolean array as a **mask** to select the elements that satisfy the condition.




````python
A = np.array([
    [1, 5, 2],
    [7, 3, 4],
    [8, 0, 2]])
print(A >= 5)
````

Similarly, you can index elements that satisfy a condition. Use `np.where` to obtain the indices of elements for which a boolean array is `True`. `np.where` returns a tuple of index arrays (one array per dimension), which can be used to index the original array. For example, to assign `111` to every element greater than or equal to `5`, do:


````python
A = np.array([
    [1, 5, 2],
    [7, 3, 4],
    [8, 0, 2]])
index = np.where(A >= 5)
A[index] = 111

print(A)
````

### Broadcasting
[Broadcasting](https://numpy.org/doc/stable/user/basics.broadcasting.html) is NumPy's way of performing element-wise operations on arrays with different shapes by automatically “stretching” dimensions of size 1 to match the other operand.
Alignment happens from the trailing axes: two axes are compatible if they are equal or one of them is 1; otherwise you'll get a ValueError.

We have already used broadcasting before, when we multiplied 1D or 2D arrays using a number, e.g.:


````python
np.array([3]) * np.ones((5,5))
````

Below, using broadcasting you can subtract the vector y from each column of x.

````python
# 2D array with shape (3,4)
X = np.array([[5, 6, 7, 8], [1, 2, 3, 4], [9, 10, 11, 12]])
print(f"X:\n{X}")
# 1D array y with shape (4,)
y = np.array([1, 2, 3, 4])
print(f"y:\n{y}")

print(f"Result of X - y:\n{X-y}")
````

Below is an example, of broadcasting which will fail due to unbroadcastable dimensions. The 2D array is of shape (4, 3), the vector (3, ), therefore subtracting them throws a ValueError.

````python
z = np.array([5, 1, 9,0])

print(f"Result of X - z:\n{X - z}")
````

Given the 2D array `A` and 1D array `b`, reshape array `b`, so that you are able to divide each column of `A` by array `b`.

````python
# Import the NumPy library
import numpy as np

# Create a 2D array x with shape (4, 3)
A = np.array([[8, 16, 24],
              [2, 4, 6],
              [12, 24, 36],
              [4, 8, 12]])

# Create a 1D array y with shape (4,)
b = np.array([2, 2, 3, 4])

# TASK: Reshape y to be a vector suitable for broadcasting
b_reshaped = b.reshape(-1,1)

# Divide each column of x by y using broadcasting
result = A / b_reshaped

# Print the original arrays and the result
print("2D Array x:\n", A)
print("1D Array y:\n", b_reshaped)
print("Result of x / y:\n", result)
````

Another operation handled by broadcasting is the outer product.
Below, arrays `u` and `v` are defined.
Performing `u @ v` computes the dot product (a scalar).
If we swap the vectors, we obtain an **outer product**, which results in a `(3,3)` 2D array.

````python
u = np.array([[6, 4, 2]])
v = np.array([[1],[2],[3]])
print(f"Dot product: {u @ v}")

print(f"Outer product:\n{v @ u}")

# of course one can also do outer product with any (1, n) array, where n is a positive integer
print(f"Outer product of (3, 1) and (1, 4) array:\n{v @ np.array([[2,4,6,8]])}")

````

You can also broadcast vectors by adding them: the following operation adds the row-shaped array `u` to the column-shaped array `v`, producing a `(3, 3)` array.

````python
u = np.array([[6, 4, 2]])
v = np.array([[1],[2],[3]])
print(f"Broadcasting using addition:\n{u + v}")
````
