### Multiplying vectors and matrices

We will start with matrix-vector multiplication. To obtain the sum of each row of an $ m\times n$ matrix, multiply the matrix on the right by an $n$-dimensional column vector of ones. The result is an $m$-dimensional column vector whose entries are the row sums.


```python
C = np.arange(6,0,-1).reshape(-1,3)
print(f"Vector containing row sums of C:\n{C @ np.ones((3,1))}")
```

Matrix–vector multiplication can be interpreted as a **linear combination of the matrix columns**, where the elements of the vector serve as the coefficients of the combination.

Below, we create a $5 \times 6$ matrix and a vector that contains ones at even indices and zeros elsewhere.  
When we multiply `H` by `v`, we obtain the **sum of the columns at even indices**, since the elements of `v` corresponding to odd indices are zero.


```python
H = np.arange(1,31).reshape(5,6)
print(f"H:\n{H}")
v = np.zeros(H.shape[1]) # vector of zeroes
v[::2] = 1 # change even indices to one
# or alternatively np.array([1, 0] *(int(H.shape[1]/2)))
print(f"H @ v:\n{H @ v}")
```

## Difference Matrix

In time series analysis — for example, when examining stock prices over a month — it is often necessary to remove **seasonality** or **trend** from the data.  
This can be achieved through **differencing**, i.e., computing the differences between consecutive values of the time series.

Below are data representing quarterly beer sales in a pub located in the city centre.  
Your task is to create a **difference matrix** (refer to the lecture slides for its structure) and use it to compute the differences between consecutive elements of the sales vector.  

You can construct the difference matrix by combining two identity (unit diagonal - np.eye() ) matrices and removing one row.

```python
quarterly_sales = np.array([20,24,30,22, 25,29,37,26, 31,33,41,31, 37,40,44,35, 43,45,49,38 ])
length = quarterly_sales.size
difference_matrix = np.eye(length)
print(difference_matrix)
```

Plot both the original and differenced vectors.  
How did the time series change after differencing?

```python
plt.plot(quarterly_sales, color = "blue", label="Original sales")
plt.plot(difference_matrix @ quarterly_sales, color = "orange", label="Differenced sales")
plt.legend()
plt.show()
```

The following matrix, `rolling_sum_matrix`, can be used to calculate the **cumulative sum** of the quarterly sales.


```python

rolling_sum_matrix = np.tril(np.ones((length,length)))
cumulative_sales = rolling_sum_matrix @ quarterly_sales
print(f"Cumulative sales: {cumulative_sales}")
```

What do you obtain if you multiply `cumulative_sales` by the transpose of the `difference_matrix`?

## Pseudoinverse of a Matrix

The **Moore–Penrose pseudoinverse** can be used to find an approximate solution to a system of **overdetermined equations** — that is, a system with more equations than unknowns.

Consider the following system of equations:
\begin{align}
-2x_1 - x_2 &= -2, \\
4x_1 - x_2 &= -8, \\
-1x_1 - x_2 &= -2.
\end{align}

When we plot each of these equations as lines, you will notice that the three lines **do not intersect at a single point**, meaning that there is **no exact solution** to this system.


```python
x1 = np.linspace(-5, 5, 1000)
x2_1 = -2*x1 + 2
x2_2 = 4*x1 + 8
x2_3 = -1*x1 + 2

plt.plot(x1, x2_1)
plt.plot(x1, x2_2)
plt.plot(x1, x2_3)
plt.xlim(-2., 1)
plt.ylim(1, 5)
plt.show()
```

We can also express the system of equations in matrix form:

\begin{align}
A\boldsymbol{x} = \boldsymbol{b}
\quad \Leftrightarrow \quad
\begin{bmatrix}
-2 & -1 \\
4 & -1 \\
-1 & -1
\end{bmatrix}
\begin{bmatrix}
x_1 \\
x_2
\end{bmatrix}
=
\begin{bmatrix}
-2 \\
-8 \\
-2
\end{bmatrix}.
\end{align}

To obtain an approximate solution for $\boldsymbol{x}$, we can use the **Moore–Penrose pseudoinverse**.  
If $A^†$ denotes the pseudoinverse of $A$, then

$$
\boldsymbol{x} = A^\dagger \boldsymbol{b}.
$$

You can compute the pseudoinverse in NumPy using the [`np.linalg.pinv`](https://numpy.org/doc/stable/reference/generated/numpy.linalg.pinv.html) function and then multiply it by $\boldsymbol{b}$ to get the least-squares solution.


```python
b = np.array([-2, -8, -2]).reshape(-1, 1)
A = np.array([[-2, -1], [4, -1], [-1, -1]])
A_plus = np.linalg.pinv(A)
print(f"A+ =\n{A_plus}")
```

We can obtain $\boldsymbol{x}$ by multiplying the pseudoinverse of $A$ by $\boldsymbol{b}$.  
Without visualization, verify whether the point lies on any of the three lines.


```python
x_res = A_plus @ b
print(x_res)
```

Below is a visualization of the point that corresponds to the solution $\boldsymbol{x}$ obtained using the pseudoinverse.

```python
plt.plot(x1, x2_1)
plt.plot(x1, x2_2)
plt.plot(x1, x2_3)
plt.xlim(-2., 1)
plt.ylim(1, 5)

plt.scatter(x_res[0], x_res[1])

plt.show()
```
