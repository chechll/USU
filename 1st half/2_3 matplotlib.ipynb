## Plotting using Package `matplotlib`

In Python, one can use package `matplotlib` to handle most plotting tasks. Its interface is very similar to plotting in Matlab. As first, we import `pyplot` and `numpy`.


```python
import numpy as np
import matplotlib.pyplot as plt
```

For an instance, this is the code for plotting a simple sine function:


```python
# We assign values from a certain range to x.
x = np.arange(0, 8, 0.1)
# We compute sin(x) for each x.
y = np.sin(x)

# We plot the resulting relationship.
plt.plot(x, y)
```

Axes can be labelled using `plt.xlabel` and `plt.ylabel`.



```python
x = np.arange(0, 8, 0.1)
y = np.sin(x)
plt.plot(x, y)

# =====
plt.xlabel("x")
plt.ylabel("y")
plt.legend(['sin(x)'])
```

We can also do other things, such as adding a grid using function `plt.grid`:



```python
x = np.arange(0, 8, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.legend(['sin(x)'])

# =====
plt.grid(ls='--')
```

We can also specify other properties of the plot when calling `plt.plot`, e.g. the line color (such as 'r' for red) or line style (such as '--' for a dashed line):



```python
x = np.arange(0, 8, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.legend(['sin(x)'])
plt.grid(ls='--')

# =====
z = np.exp(0.1*x)
plt.plot(x, z, 'r--')
```

We update the legend  expression `np.exp` can be written using LaTeX notation, in which case it displays in the proper mathematical format:



```python
x = np.arange(0, 8, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.legend(['sin(x)'])
plt.grid(ls='--')
z = np.exp(0.1*x)
plt.plot(x, z, 'r--')

# =====
plt.legend(['$\sin(x)$', '$e^{0.1x}$'])
```

## Scatter plots

Below is an example of a 4-dimensional visualisation using a scatter plot. We also add a simple the legend to the plot which guids the scales of the circles.

```python
np.random.seed(14)


N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
# area = (30 * np.random.rand(N))**2  # 0 to 15 point radii
area = np.random.randint(5,12,N)**3  # 0 to 15 point radii

for circle_diam in [5**3, 8**3, 11**3]:
    plt.scatter([], [], c='k', alpha=0.3, s=circle_diam,
                label=str(circle_diam))
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='Diameter')


plt.scatter(x, y, s=area, c=colors, alpha=0.5, label = area)

plt.show()
```
